```
from pyspark.sql import SparkSession
from pyspark.sql.functions import col,lit
```

```
spark = SparkSession.builder.appName("merge dataframe").getOrCreate()
input1_df = spark.read.option("delimiter","|").csv("/FileStore/tables/input1_csv.txt",header=True)
input2_df = spark.read.option("delimiter","|").csv("/FileStore/tables/input2_csv.txt",header=True)

input1_df.show()
input2_df.show()

df_add = input1_df.withColumn("gender",lit("null"))
df_add.show()
df_add.union(input2_df).show()
```

```
from pyspark.sql.types import *
schema = StructType(
    [
        StructField("Name",StringType(),True),
        StructField("Age", StringType(),True),
        StructField("Gender",StringType(),True),
    ]
)
```

```
input3_df = spark.read.option("delimiter","|").csv("/FileStore/tables/input1_csv.txt",header=True,schema = schema)
input4_df = spark.read.option("delimiter","|").csv("/FileStore/tables/input2_csv.txt",header=True,schema = schema)
```

```
df_union = input3_df.union(input4_df).show()
```
